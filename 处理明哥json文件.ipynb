```python
import requests
url_100 = 'http://106.13.151.200:3334'
url_300 = 'http://106.13.151.200:3335'
headers = {'Content-Type':'application/x-www-form-urlencoded'}
data = {'content':'网传NASA 称因地球重力角度，2 月 11 日是唯一能让扫把独立站立的日子，是真的吗？有科学依据吗？'}
r = requests.post(url_100, headers=headers, data=data)
```

```python
import pandas as pd
import numpy as np
```

```python
df_all = pd.DataFrame(a)
```

```python
for i in range(2133,len(a)):
    if df_all['content'][i] != '':
        data = {'content':df_all['content'][i]}
        result1 = requests.post(url_100, headers=headers, data=data).json()['lv1_tag']
        result2 = requests.post(url_300, headers=headers, data=data).json()['lv1_tag']
        df_all['100_cat'][i] = result1['tag']
        df_all['100_p'][i] = result1['score']
        df_all['300_cat'][i] = result2['tag']
        df_all['300_p'][i] = result2['score']
```
